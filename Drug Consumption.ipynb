##### Change Age, Gender, Country, Ethnicity, Education for EDA and rename Columns

```python
df = df.rename(columns={'Nscore': 'Neuroticism', 
                        'Escore': 'Extraversion', 
                        'Oscore': 'Openness', 
                        'Ascore': 'Agreeableness', 
                        'Cscore': 'Conscientiousness', 
                        'Impulsive': 'Impulsiveness', 
                        'SS': 'Sensations'})
```

```python
df.dtypes
```

```python
# ID
# will not be used in Model
df = df.drop('ID', axis=1)
```

```python
df.loc[df.Age == -0.95197, "Age"]= '(18-24)'
df.loc[df.Age == -0.07854, "Age"]= '(25-34)'
df.loc[df.Age == 0.49788, "Age"]= '(35-44)'
df.loc[df.Age == 1.09449, "Age"]= '(45-54)'
df.loc[df.Age == 1.82213, "Age"]= '(55-64)'
df.loc[df.Age == 2.59171, "Age"]= '(65+)'
```

```python
df.Gender = ['male' if i == -0.48246 else 'female' for i in df.Gender]
```

```python
df.loc[df.Country == 0.96082, "Country"]= 'UK'
df.loc[df.Country == -0.57009, "Country"]= 'USA'
df.loc[df.Country == -0.28519, "Country"]= 'Other'
df.loc[df.Country == 0.24923, "Country"]= 'Canada'
df.loc[df.Country == -0.09765, "Country"]= 'Australia'
df.loc[df.Country == 0.21128, "Country"]= 'Ireland'
df.loc[df.Country == -0.46841, "Country"]= 'NewZealand'
```

```python
df.loc[df.Ethnicity == -0.31685, "Ethnicity"]= 'White'
df.loc[df.Ethnicity == 0.11440, "Ethnicity"]= 'Other'
df.loc[df.Ethnicity == -1.10702, "Ethnicity"]= 'Black'
df.loc[df.Ethnicity == -0.50212, "Ethnicity"]= 'Asian'
df.loc[df.Ethnicity == -0.22166, "Ethnicity"]= 'Mixed-White/Black'
df.loc[df.Ethnicity == 0.12600, "Ethnicity"]= 'Mixed-White/Asian'
df.loc[df.Ethnicity == 1.90725, "Ethnicity"]= 'Mixed-Black/Asian'
```

```python
# Asumption: starts school at six years
df.loc[df.Education == -0.61113, "Education"]= 'Undergrad'
df.loc[df.Education == 0.45468, "Education"]= 'Uni'
df.loc[df.Education == 1.16365, "Education"]= 'Masters'
df.loc[df.Education == -0.05921, "Education"]= 'Professional'
df.loc[df.Education == -1.22751, "Education"]= '12'
df.loc[df.Education == -1.73790, "Education"]= '10'
df.loc[df.Education == 1.98437, "Education"]= 'PHD'
df.loc[df.Education == -1.43719, "Education"]= '11'
df.loc[df.Education == -2.43591, "Education"]= 'Under 10'
```
