```python
# People who consume Semeron can't be trusted ('Over-claimers'). Drop them. 
# After that, Sem has no value, drop col
df[(df.Semer != 'CL0')].shape
```

```python
df.drop(df[(df.Semer != 'CL0')].index, axis=0, inplace=True)
```

```python
# Rename Categories for Use 
drugs = ['Alcohol', 'Amphet', 'Amyl', 'Benzos', 'Caff', 'Cannabis', 'Choc', 'Coke', 'Crack', 'Ecstasy', 'Heroin',
'Ketamine', 'Legalh', 'LSD', 'Meth', 'Shrooms', 'Nicotine', 'Semer', 'VSA']
```

```python
for i in drugs:
    df.loc[df[i] == "CL0", i]= 'Never'
    df.loc[df[i] == "CL1", i]= 'Over Decade ago'
    df.loc[df[i] == "CL2", i]= 'In last Decade'
    df.loc[df[i] == "CL3", i]= 'Last Year'
    df.loc[df[i] == "CL4", i]= 'Last Month'
    df.loc[df[i] == "CL5", i]= 'Last Week'
    df.loc[df[i] == "CL6", i]= 'Last Day'
```

```python
for i in drugs:
    df[i] = df[i].astype('category')
```
